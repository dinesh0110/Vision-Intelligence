# Lab Exercise 5: Geometric Transformations and Affine Transformations 
•	Objective: Apply geometric transformations and affine transformations to images. 

•	Task: Perform image scaling, rotation, translation, and affine transformations. Visualize the effect of each transformation on an image. 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load the image
image = cv2.imread("./image2.jpg")
print("Original Image shape:", image.shape)

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Scale the image by 50% (reduce size)
scale_factor = 0.5
scaled_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)

# Display the scaled image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
plt.title("Scaled Image (50%)")
plt.axis('off')
plt.show()
```

```python
# Define translation matrix to shift the image by 50 pixels right and 30 pixels down
tx, ty = 50, 30
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Display the translated image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated Image (50 px right, 30 px down)")
plt.axis('off')
plt.show()
```

```python
# Define the rotation center and angle
center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 45  # Rotate by 45 degrees
scale = 1.0  # No scaling

# Create the rotation matrix and apply it
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

# Display the rotated image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image (45°)")
plt.axis('off')
plt.show()
```

```python
# Define points for affine transformation
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

# Get the affine transformation matrix and apply it
affine_matrix = cv2.getAffineTransform(pts1, pts2)
affine_transformed_image = cv2.warpAffine(image, affine_matrix, (image.shape[1], image.shape[0]))

# Display the affine transformed image
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(affine_transformed_image, cv2.COLOR_BGR2RGB))
plt.title("Affine Transformed Image")
plt.axis('off')
plt.show()
```

# Lab Exercise 5: Geometric Transformations and Affine Transformations

## Objective
This lab explores how to perform various geometric transformations, including scaling, rotation, translation, and affine transformations, and visualize their impact on an image.

## Transformations Applied

1. **Scaling**:
   - Resizes the image based on a specified scale factor.
   - In this exercise, the image was scaled to 50% of its original size, effectively reducing its dimensions.

2. **Translation**:
   - Shifts the image along the X and Y axes.
   - The example shifts the image 50 pixels to the right and 30 pixels downward, moving the image within the frame.

3. **Rotation**:
   - Rotates the image by a specified angle around a central point.
   - In this lab, the image was rotated by 45° around its center, which changes its orientation while preserving shape and scale.

4. **Affine Transformation**:
   - Uses a linear mapping based on three source and destination points to transform the image.
   - This technique allows for various effects, such as skewing and rotation, while maintaining parallelism of lines.

## Summary of Results
Each transformation uniquely alters the image:
- **Scaling** changes the size without affecting the shape or content.
- **Translation** shifts the image position but retains all image details.
- **Rotation** adjusts the orientation, useful for aligning images at different angles.
- **Affine Transformation** provides flexibility for complex shifts, combining translation, scaling, and rotation effects.

## Conclusion
Geometric transformations, including affine transformations, are essential tools for image manipulation. They are foundational in image processing and computer vision, enabling image alignment, orientation adjustments, and preparation for further analysis. Each transformation offers specific advantages based on the desired effect, providing versatility in manipulating image data.
