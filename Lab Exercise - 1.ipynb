```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(img):
    cv.imshow("Image", img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
```

```python
# original image
v1 = cv.imread("./location.jpg")  
imshow(v1)
plt.imshow(v1)  
```

```python
v2 = cv.cvtColor(v1, cv.COLOR_BGR2RGB)  # Convert BGR to RGB
imshow(v2) 
plt.imshow(v2)
```

```python
# Gray Conversion
v3 = cv.imread("./location.jpg")  
imshow(v3) 
plt.imshow(v3)
v3GRAY = cv.cvtColor(v1, cv.COLOR_BGR2GRAY)
imshow(v3GRAY)
plt.imshow(v3GRAY)
```

```python
# HSV Conversion

v4 = cv.cvtColor(v1, cv.COLOR_BGR2HSV)  
imshow(v4)
plt.imshow(v4)
```

```python
# YUV Conversion
v5 = cv.cvtColor(v1, cv.COLOR_BGR2YUV)  
imshow(v5)
plt.imshow(v5)
```
