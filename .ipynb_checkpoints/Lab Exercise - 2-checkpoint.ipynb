# Lab Exercise 2: Gaussian and Laplacian Pyramids 
•	Objective: Learn how to build multi-resolution pyramids for an image. 


•	Task: Create Gaussian and Laplacian pyramids for a given image, then use these pyramids to perform image blending between two images. 

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt
````

````python
# Load the images (replace 'image1.jpg' and 'image2.jpg' with your image paths)
img1 = cv2.imread("./image1.jpg")
img2 = cv2.imread("./image2.jpg")

# Resize images to the same size if necessary
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Convert from BGR to RGB for displaying with matplotlib
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Display the images
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(img1_rgb)
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2_rgb)
plt.title("Image 2")
plt.axis('off')
plt.show()
````

````python
# Function to create a Gaussian pyramid
def gaussian_pyramid(image, levels):
    pyramid = [image]
    for i in range(levels):
        image = cv2.pyrDown(image)
        pyramid.append(image)
    return pyramid

# Set the number of levels for the pyramid
levels = 5

# Create Gaussian pyramids for both images
gaussian_pyramid_img1 = gaussian_pyramid(img1, levels)
gaussian_pyramid_img2 = gaussian_pyramid(img2, levels)
````

````python
# Function to create a Laplacian pyramid
def laplacian_pyramid(gaussian_pyramid):
    pyramid = [gaussian_pyramid[-1]]
    for i in range(len(gaussian_pyramid) - 1, 0, -1):
        size = (gaussian_pyramid[i - 1].shape[1], gaussian_pyramid[i - 1].shape[0])
        gaussian_expanded = cv2.pyrUp(gaussian_pyramid[i], dstsize=size)
        laplacian = cv2.subtract(gaussian_pyramid[i - 1], gaussian_expanded)
        pyramid.append(laplacian)
    return pyramid

# Create Laplacian pyramids for both images
laplacian_pyramid_img1 = laplacian_pyramid(gaussian_pyramid_img1)
laplacian_pyramid_img2 = laplacian_pyramid(gaussian_pyramid_img2)
````

````python
# Blend the Laplacian pyramids
laplacian_blended = []
for lap1, lap2 in zip(laplacian_pyramid_img1, laplacian_pyramid_img2):
    rows, cols, _ = lap1.shape
    laplacian_blend = np.hstack((lap1[:, :cols // 2], lap2[:, cols // 2:]))
    laplacian_blended.append(laplacian_blend)
````

````python
# Start with the smallest image in the Laplacian pyramid
blended_image = laplacian_blended[0]

# Reconstruct the image by progressively adding each level of the pyramid
for i in range(1, levels + 1):
    # Upscale the current blended image to the next level's size
    blended_image = cv2.pyrUp(blended_image)
    
    # Resize if necessary to match the size of the current level in laplacian_blended
    if blended_image.shape[:2] != laplacian_blended[i].shape[:2]:
        blended_image = cv2.resize(blended_image, (laplacian_blended[i].shape[1], laplacian_blended[i].shape[0]))

    # Add the current level's Laplacian image to reconstruct details
    blended_image = cv2.add(blended_image, laplacian_blended[i])

# Convert the final blended image to RGB format for matplotlib display
blended_image_rgb = cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB)

# Display the final blended image using matplotlib
plt.imshow(blended_image_rgb)
plt.title("Blended Image")
plt.axis('off')
plt.show()
````

# Lab Exercise 2: Gaussian and Laplacian Pyramids

## Objective
Learn how to build multi-resolution pyramids (Gaussian and Laplacian) for image processing and use them for seamless image blending.

## Steps and Observations

1. **Image Loading**:
   - Two images were loaded and resized to the same dimensions for blending.

2. **Gaussian Pyramid**:
   - Each image was downsampled multiple times to create a Gaussian pyramid.
   - Gaussian pyramids help in obtaining lower-resolution versions of an image.

3. **Laplacian Pyramid**:
   - The Laplacian pyramid was created by subtracting the Gaussian pyramid levels.
   - The Laplacian pyramid stores edge and detail information at each level.

4. **Image Blending**:
   - The Laplacian pyramids of both images were blended by taking the left half of one image and the right half of the other at each level.
   - This step allows for a smooth transition between the two images.

5. **Reconstruction**:
   - The blended Laplacian pyramid was used to reconstruct a single blended image.
   - This final image has a seamless transition between the two original images.

## Usage

Run the provided code to visualize the Gaussian and Laplacian pyramids and the final blended result. The images display in matplotlib windows for better clarity.
