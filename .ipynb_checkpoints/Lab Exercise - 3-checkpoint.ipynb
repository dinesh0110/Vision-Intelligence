# Lab Exercise 3: Feature Detection and Matching 
•	Objective: Implement feature detection algorithms and matching between images. 


•	Task: Use SIFT or SURF to detect features in two images and apply feature matching techniques like brute-force matching or FLANN. 


````python
import cv2
import numpy as np
from matplotlib import pyplot as plt
````

````python
# Load the images
img1 = cv2.imread("./image1_crop1.jpg", cv2.IMREAD_GRAYSCALE)  # Query image
img2 = cv2.imread("./image2_crop2.jpg", cv2.IMREAD_GRAYSCALE)  # Train image

# Check and print the shapes of the images
print("Image 1 shape:", img1.shape)
print("Image 2 shape:", img2.shape)

# Resize images to the same size if necessary
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
    print("Resized Image 2 to match Image 1 shape:", img2.shape)
````

````python
# Initialize SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

# Display the number of features detected
print("Number of keypoints in Image 1:", len(keypoints1))
print("Number of keypoints in Image 2:", len(keypoints2))
````

````python
# Initialize Brute-Force Matcher
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

# Match descriptors
matches = bf.match(descriptors1, descriptors2)

# Sort matches by distance (best matches first)
matches = sorted(matches, key=lambda x: x.distance)

# Display the number of matches
print("Number of matches found:", len(matches))
````

````python
# Draw the matches
match_img = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches[:50], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Display the matching result
plt.figure(figsize=(15, 10))
plt.imshow(match_img, cmap='gray')
plt.title("Feature Matching between Image 1 and Image 2")
plt.axis('off')
plt.show()
````

# Lab Exercise 3: Feature Detection and Matching

## Objective
Implement feature detection algorithms (SIFT) and feature matching (Brute-Force and FLANN) to match features between two images.

## Steps and Observations

1. **Image Loading and Shape Checking**:
   - Two grayscale images are loaded, and their shapes are checked to ensure compatibility.
   - If necessary, images are resized to match each other's dimensions.

2. **Feature Detection using SIFT**:
   - SIFT detects keypoints and computes descriptors for each image.
   - The number of detected keypoints provides an estimate of the distinct features found in each image.

3. **Brute-Force Matching**:
   - Brute-Force matching is applied to compare descriptors from both images.
   - Matches are sorted by distance, with closer matches representing better matches.
   - Visualized the top matches, showing connected lines between matching points.

4. **FLANN-based Matching (Optional)**:
   - FLANN-based matcher is used as an alternative for larger datasets, using approximate nearest neighbors.
   - A ratio test is applied to filter out weak matches, resulting in "good matches" for better accuracy.

## Usage

Run the code to detect features in two images and visualize the feature matches. The images display in matplotlib windows for clear visualization of feature matching.
